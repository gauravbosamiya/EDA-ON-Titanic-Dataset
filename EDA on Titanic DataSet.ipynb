```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# 0 = No survived
# 1 = Yes survived
```

```python
df = pd.read_csv('DataSet/titanic.csv')
df.head()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df = df.drop(df[['Cabin']],axis=1)
```

```python
plt.figure(figsize=(9,15))
fig_dimns = (3,2)

plt.subplot2grid(fig_dimns, (0,0))
sur = df['Survived'].value_counts().plot(kind='bar',title='Death vs Survival count',color='maroon')
for bars in sur.containers:
    sur.bar_label(bars)

plt.subplot2grid(fig_dimns, (0,1))
pclass = df['Pclass'].value_counts().plot(kind='bar',title='Passanger class count',color='maroon')
for bars in pclass.containers:
    pclass.bar_label(bars)

plt.subplot2grid(fig_dimns, (1,0))
sex = df['Sex'].value_counts().plot(kind='bar',title='Gender Count',color='maroon')
for bars in sex.containers:
    sex.bar_label(bars)

plt.subplot2grid(fig_dimns, (1,1))
emb = df['Embarked'].value_counts().plot(kind='bar',title='Embarked Count',color='maroon')
for bars in emb.containers:
    emb.bar_label(bars)
```

```python
ax=sns.countplot(data=df,x='Sex',hue='Pclass',palette = 'inferno')
plt.title('Pclass vs Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```
